# Building a simple dataset

```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels = []
train_samples = []
```

```python
# example data:
#    an experienced drug was tested on individuals from 13 to 100 in a clinical trial
#    the trial had 2100 participants. half were under 65 years old. half were 65 or older
#    95% of patients 65 or older experience side effects
#    95% of patients under 65 experienced no side effects
    
```

```python
for i in range(50):
    #the 5% of younger individuals who did experience side effects
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1)
    
    #the 5% of older individuals who did not experience side effects
    random_older = randint(65, 100)
    train_samples.append(random_older)
    train_labels.append(0)
    
for i in range(1000):
    #the 95% of younger individuals who did not experience side effects
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)
    
    #the 95% of older individuals who did experience side effects
    random_older = randint(65, 100)
    train_samples.append(random_older)
    train_labels.append(1)
    
    
    
```

```python
for i in train_samples:
    print(i)
```

```python
for i in train_labels:
    print(i)
```

```python
#convert train samples and labels to np arrays to be able to use in fit function documentation
train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
train_labels, train_samples = shuffle(train_labels, train_samples)
```

```python
#now we scale down the data from the original range of between 13 to 100 to a scale of 0 to 1
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1))
```

```python
#print scaled data
for i in scaled_train_samples:
    print(i)
```

# Simple tf.keras Sequential Model

```python
model = Sequential([
    Dense(units=16, input_shape=(1,), activation='relu'), #this is the first hidden layer
    Dense(units=32, activation='relu'), #second hidden layer
    Dense(units=2, activation= 'softmax') #output layer
])
```
